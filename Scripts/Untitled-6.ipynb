```python
import pandas as pd

claims = [
    "claim_id=1001|patient=Alice|dx=E11.9|amount=1200.50|date=2026-01-05",
    "claim_id=1002|patient=Bob|dx=I10|amount=800|date=01/06/2026",
    "claim_id=1003|patient=Charlie|dx=E11.9|amount=950.75|date=2026/01/07",
    "claim_id=1004|patient=David|dx=J45.909|amount=|date=2026-01-08",
    "claim_id=1005|patient=Alice|dx=E11.9|amount=1500|date=2026-01-09",
    "claim_id=1006|patient=Bob|dx=I10|amount=700.25|date=bad_date",
    "claim_id=1007|patient=Eve|dx=Z00.00|amount=100|date=2026-01-10"
]
```

```python
clm_lst = []
ptnt_lst = []
dx_lst = []
amount_lst = []
date_lst = []

for claim in claims:
    clm_lst.append(claim.split('|')[0].split('=')[1])
    ptnt_lst.append(claim.split('|')[1].split('=')[1])
    dx_lst.append(claim.split('|')[2].split('=')[1])
    amount_lst.append(claim.split('|')[3].split('=')[1])
    date_lst.append(claim.split('|')[4].split('=')[1])
    
df = pd.DataFrame({
    'claim_id': clm_lst,
    'patient' : ptnt_lst,
    'dx' : dx_lst,
    'amount' : amount_lst,
    'date' : date_lst
})

df
```

```python
df['date'] = pd.to_datetime(df['date'], format = 'mixed', errors = 'coerce')
df
```

```python
df['amount'] = pd.to_numeric(df['amount'], errors = 'coerce')

df['amount'].groupby(df['patient']).sum()  
```
