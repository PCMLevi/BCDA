```python
s = "ID-2026-0001"
letters = []
digits = []
partitions = []
for i in s:
    if i.isdigit():
        digits.append(i)
    elif i.isalpha():
        letters.append(i)
    else:
        partitions.append(i)
        
letters = ''.join(letters).lower()
digits = ''.join(digits)
partitions = set(''.join(partitions))
print(letters, digits,partitions)
```
